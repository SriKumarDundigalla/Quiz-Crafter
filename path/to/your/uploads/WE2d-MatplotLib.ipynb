```python
# scatter plot with different colors and sizes and random data

import random

x = [random.randint(0, 100) for i in range(10)]
y = [random.randint(0, 100) for i in range(10)]
sizes = [random.randint(0, 100) for i in range(10)]
kcolors = ['red' if value > 50 else 'blue' for value in y]

plt.scatter(x, y, label='Some Data', color=kcolors, s=sizes, marker="o")
# line at y = 50
plt.axhline(y=50, color='r', linestyle='--')
# line y = 2x + 3
y2 = [2*x + 3 for x in x]
plt.plot(x, y2, linestyle='--', lw=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```
